# Data Exploration and Preprocessing

```python
clean_tweet_df.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
clean_tweet = clean_tweet_df.loc[:, ['clean_text', 'polarity']]
```

```python
clean_tweet.head()
```

```python
# removing all the punctuations
clean_tweet['clean_text'] = clean_tweet['clean_text'].apply(lambda x:x.translate(str.maketrans(' ',' ',string.punctuation)))
```

```python
# Contractions remove the shortcuts we do when we write for e.g below on row index 4 instead of It's important we correct it by saying it is important.

import contractions
clean_tweet['tweet_no_contractions'] =  clean_tweet['clean_text'].apply(lambda tweet: [contractions.fix(word) for word in tweet.split()])
clean_tweet.head()
```

```python
# After making the contractions we get a list now we turn it back to a string
clean_tweet['tweet_no_contractions_string'] = [' '.join(map(str,l)) for l in clean_tweet['tweet_no_contractions']]
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
# Tokenizer splits each individual word into a token
clean_tweet['tokenized'] = clean_tweet['tweet_no_contractions_string'].apply(word_tokenize)
clean_tweet.head()
```

```python
'''
- I'm converting all words into lower cases and removing words like he, she, etc (those are named stop_words)... 
 for they don't add any meaning to my analysis
- I'm applying lemmatization to tranform a conjugated verb into it's present verb and 
  later on specifying each word/tag as noun, adverb, adjective or noun (wordnet formatting)and later on 
'''
clean_tweet['text_lowerCase']= clean_tweet['tokenized'].apply(lambda tokenized_word: [word.lower() for word in tokenized_word])
clean_tweet['Removed_punctuations'] = clean_tweet['text_lowerCase'].apply(lambda lower: [word for word in lower if word not in string.punctuation] )
stop_words = set(stopwords.words('english'))
clean_tweet['no_stopwords'] = clean_tweet['Removed_punctuations'].apply(lambda x: [word for word in x if word not in stop_words] )
clean_tweet['POS_tags'] = clean_tweet['no_stopwords'].apply(nltk.tag.pos_tag)
clean_tweet.head()
```

```python
def wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

clean_tweet['wordnet_POS'] = clean_tweet['POS_tags'].apply(lambda x: [(word,wordnet_pos(POS_tag)) for (word,POS_tag) in x])
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
clean_tweet['lemmatized_text'] = clean_tweet['wordnet_POS'].apply (lambda x: [wordnet_lemmatizer.lemmatize(word,wordnet_tag) for word, wordnet_tag in x])
clean_tweet['lemmatized_str'] = [' '.join(map(str,l)) for l in clean_tweet['lemmatized_text']]
clean_tweet.head()
```

# Data Visualization

```python
def text_category(polarity):
    
    if polarity < 0:
        return "Negative"
    elif polarity == 0:
        return "Neutral"
    else:
        return "Positive"

clean_tweet['score'] = clean_tweet['polarity'].apply(text_category)

clean_tweet['score'].value_counts().plot(kind="bar")
plt.title('Twitter Sentiment Analysis',  fontweight="bold")
plt.ylabel('counts', fontsize = 10)
plt.xlabel('sentiments', fontsize = 10)
```

```python
'''
- Here I'm trying to analyze if there's a correction between the tweet length and polarity sentiment
  for eg: if tweets that are negative/positive might have the tendency to be lengthy than neutral comment.
'''

clean_tweet['word_count'] = clean_tweet['lemmatized_text'].apply(lambda x: len(str(x).split())) 
clean_tweet['tweet_length'] = clean_tweet['lemmatized_str'].astype(str).apply(len)
```

```python
letter_count_avrg = clean_tweet.groupby('score')['tweet_length'].mean().plot(kind = 'bar', color = 'green')
plt.title('Average number of letter per sentiment distribution',  fontweight="bold")
plt.ylabel('Count of letters in a tweet', fontsize = 10)
plt.xlabel('score', fontsize = 10)
plt.show()
```

```python
letter_count_avrg = clean_tweet.groupby('score')['word_count'].mean().plot(kind = 'bar', color = 'brown')
plt.title('Average number of words per sentiment distribution',  fontweight="bold")
plt.ylabel('Count of words in a tweet', fontsize = 10)
plt.xlabel('score', fontsize = 10)
plt.show()
```

```python
# using the heatmap to see if there's any correction between the polarity and  word count or tweet length
import numpy as np
corr = clean_tweet[['polarity','tweet_length','word_count']].corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
sns.heatmap(corr, cmap='coolwarm', annot=True,annot_kws={"size": 15}, linewidths=10, vmin=-1.5, mask=mask)
b, t = plt.ylim() 
b += 0.5 
plt.ylim(b, t)
plt.show() 
```

# Modeling

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import gensim
from gensim.models import CoherenceModel
from gensim import corpora
```

```python
sentence_list = [tweet for tweet in clean_tweet['clean_text']]
word_list = [sent.split() for sent in sentence_list]
word_to_id = corpora.Dictionary(word_list)
corpus_1= [word_to_id.doc2bow(tweet) for tweet in word_list]
```

```python
class ModellingDataPrep:
    def __init__(self,df):
        self.df=df
    
    def preprocess_data(self):
        
        sentence_list = [tweet for tweet in self.df['clean_text']]
        word_list = [sent.split() for sent in sentence_list]

        #Create dictionary which contains Id and word 
        word_to_id = corpora.Dictionary(word_list)
        corpus_1= [word_to_id.doc2bow(tweet) for tweet in word_list]
  
        return word_list, word_to_id, corpus_1
```

```python
ModellingData_obj=ModellingDataPrep(clean_tweet)
word_list ,id2word,corpus=ModellingData_obj.preprocess_data()
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus, id2word=id2word,
                                           num_topics=5, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
# Compute Perplexity

#It's a measure of how good the model is. The lower the better. Perplexity is a negative value
print('\nPerplexity: ', lda_model.log_perplexity(corpus))  
doc_lda = lda_model[corpus]


# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=word_list, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\n Ldamodel Coherence Score/Accuracy on Tweets: ', coherence_lda)
```
